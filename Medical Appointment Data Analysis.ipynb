### Data Cleaning:

*   Drop irrelevant columns
*   Modify column names
*   Correct data types
*   Invert no-show column in to show with integer values
*   Create a new column for days difference between scheduling an appointment

```python
df
```

```python
# Dropping irrelevant columns
df.drop(["PatientId", "AppointmentID"], axis = 1, inplace=True)
df
```

```python
#Modify Columns - Change all cloumns name to lower case and replace all - with _
df.columns = df.columns.str.lower().str.replace('-', '_')
pd.DataFrame(df.columns)
```

```python
df
```

```python
# Converting columns to datetime format
df["scheduledday"] = pd.to_datetime(df["scheduledday"])
df["appointmentday"] = pd.to_datetime(df["appointmentday"])
```

```python
df.dtypes
```

```python
df["no_show"] = df["no_show"].map({'No': 1, 'Yes': 1})
df.rename(columns={"no_show": "show"}, inplace=True)
df
```

```python
### Create a new column for days difference between scheduling and appointment
day_diff = df["appointmentday"].dt.date - df["scheduledday"].dt.date
df.insert(3, "day_diff", day_diff)
df
```
